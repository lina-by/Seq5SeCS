```python
import re
from collections import defaultdict

word_frequency = defaultdict(lambda: defaultdict(int))

def tokenize(text):
    # Remove punctuation and split by whitespace
    return re.findall(r'\b\w+\b', text.lower())
```

## Nave Bayes

```python
from collections import defaultdict

word_frequency = defaultdict(lambda: defaultdict(int))

for i, row in train_s.iterrows():
    lang = row['Label']
    words = tokenize(row['Text'])
    
    for word in words:
        word_frequency[lang][word] += 1

if 'en' in word_frequency:
    top_10_words = sorted(word_frequency['en'].items(), key=lambda x: x[1], reverse=True)[:10]
    print("Top 10 words in English:", top_10_words)
```

```python
import math
from collections import defaultdict
from sklearn.metrics import accuracy_score, classification_report

# Laplace smoothing constant
alpha = 1

# Calculate prior probabilities P(L)
def calculate_language_priors(word_frequency):
    total_sentences = sum([sum(lang_counts.values()) for lang_counts in word_frequency.values()])
    priors = {}
    for lang, word_counts in word_frequency.items():
        priors[lang] = sum(word_counts.values()) / total_sentences  # Prior for each language
    return priors

# Calculate conditional probabilities P(w_i | L) with Laplace smoothing
def calculate_conditional_probabilities(word_frequency, alpha=1):
    word_probs = defaultdict(lambda: defaultdict(float))  # {language: {word: prob}}
    
    for lang, word_counts in word_frequency.items():
        total_words_in_lang = sum(word_counts.values())  # Total number of words in this language
        vocab_size = len(word_counts)  # Size of the vocabulary for the language
        for word, count in word_counts.items():
            # Calculate conditional probability P(w | L) for each word in the language
            word_probs[lang][word] = (count + alpha) / (total_words_in_lang + alpha * vocab_size)
    return word_probs

# Nave Bayes prediction for each sentence
def predict_language(sentence, priors, word_probs, word_frequency, alpha=1):
    words = tokenize(sentence)  # Tokenize the sentence into words
    best_lang = None
    best_prob = -float('inf')
    
    for lang in priors:
        log_prob = math.log(priors[lang])  # Start with the prior probability log(P(L))
        
        # Add log of conditional probabilities for each word in the sentence
        for word in words:
            if word in word_probs[lang]:
                log_prob += math.log(word_probs[lang][word])
            else:
                # If the word is not in training data for this language, apply Laplace smoothing
                total_words_in_lang = sum(word_frequency[lang].values())  # Total words in language
                vocab_size = len(word_frequency[lang])  # Vocabulary size for the language
                log_prob += math.log(alpha / (total_words_in_lang + alpha * vocab_size))
        
        # Keep track of the language with the highest log probability
        if log_prob > best_prob:
            best_prob = log_prob
            best_lang = lang
    
    return best_lang

# Evaluate the model
def evaluate_naive_bayes(train_s, validation_s):
    # 1. Calculate priors
    priors = calculate_language_priors(word_frequency)

    # 2. Calculate conditional probabilities P(w | L)
    word_probs = calculate_conditional_probabilities(word_frequency, alpha)

    # 3. Predict for validation set
    predictions = [predict_language(sentence, priors, word_probs, word_frequency) for sentence in tqdm(validation_s["Text"])]

    # 4. Evaluate accuracy
    print(f"Nave Bayes Accuracy: {accuracy_score(validation_s['Label'], predictions):.4f}")
    print(classification_report(validation_s["Label"], predictions))

# Call the evaluation function
evaluate_naive_bayes(train_s, validation_s)
```

# LSTM

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split

# Define a dataset class
class TextDataset(Dataset):
    def __init__(self, texts, labels, vocab):
        self.texts = texts
        self.labels = labels
        self.vocab = vocab
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        # Tokenize on a per-character basis
        tokens = list(self.texts[idx])  # Each character is a token
        indices = [self.vocab.get(char, self.vocab["<unk>"]) for char in tokens]  # Handle unknown characters
        # Ensure labels are integers
        label = int(self.labels[idx]) if isinstance(self.labels[idx], str) else self.labels[idx]
        return torch.tensor(indices, dtype=torch.long), label
```

```python
def build_vocab(texts):
    vocab = {"<pad>": 0, "<unk>": 1}  # Add padding and unknown token
    for text in texts:
        for char in text:
            if char not in vocab:
                vocab[char] = len(vocab)
    return vocab
```

```python
def collate_fn(batch):
    # Find the maximum sequence length in the batch
    max_len = max([len(item[0]) for item in batch])
    
    # Pad all sequences to the same length (max_len)
    padded_texts = [torch.cat([item[0], torch.zeros(max_len - len(item[0]), dtype=torch.long)]) for item in batch]
    
    # Extract labels
    labels = torch.tensor([item[1] for item in batch], dtype=torch.long)
    
    return torch.stack(padded_texts, dim=0), labels
```

```python
# Train-validation split
train_s, validation_s = train_test_split(
    main_s, 
    test_size=0.2, 
    #stratify=main_s["Label"], 
    random_state=42
)
```

```python
# Build vocab based on the training set texts
vocab = build_vocab(train_s["Text"].tolist())

# Convert labels to indices
label_to_index = {label: idx for idx, label in enumerate(set(main_s["Label"]))}
main_s["Label"] = main_s["Label"].map(label_to_index)

# Create Datasets and DataLoaders
train_dataset = TextDataset(train_s["Text"].tolist(), train_s["Label"].tolist(), vocab)
val_dataset = TextDataset(validation_s["Text"].tolist(), validation_s["Label"].tolist(), vocab)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)
```

```python
# Define LSTM model
class LSTMLanguageClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, output_dim, num_layers=1):
        super(LSTMLanguageClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=vocab["<pad>"])
        self.lstm = nn.LSTM(embed_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        embedded = self.embedding(x)
        _, (hidden, _) = self.lstm(embedded)
        out = self.fc(hidden[-1])
        return out
```

```python
# Model initialization
vocab_size = len(vocab)
embed_dim = 100
hidden_dim = 64
output_dim = len(label_to_index)
num_layers = 1

# Set device to CPU (if it's not already) to avoid using a GPU if it's slower
device = torch.device("cpu")  # Force the use of CPU if the GPU is not available or slower
model = LSTMLanguageClassifier(vocab_size, embed_dim, hidden_dim, output_dim, num_layers).to(device)

# Reduce batch size and epochs
batch_size = 16  # Smaller batch size
epochs = 5  # Reduced epochs for quicker training

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop (with reduced epochs and batch size)
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for texts, labels in train_loader:
        texts, labels = texts.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(texts)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss / len(train_loader):.4f}")
```

```python
# Evaluation loop (with no_grad to save memory)
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for texts, labels in val_loader:
        texts, labels = texts.to(device), labels.to(device)
        outputs = model(texts)
        predictions = torch.argmax(outputs, dim=1)
        correct += (predictions == labels).sum().item()
        total += labels.size(0)

print(f"Validation Accuracy: {correct / total:.4f}")
```
